```python
days = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']
accident_table = data.groupby(['Crash year', 'Day of the week']).size()
accident_table = accident_table.rename_axis(['Year', 'Weekday'])\
                               .unstack('Weekday')\
                               .reindex(columns=days)
```

```python
accident_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(accident_table, cmap='RdYlGn_r',annot=True)
plt.title('\nAccidents by Years and Weekdays\n', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('');
```

```python
monthly = data[['Month of year', 'Crash identifier']].groupby(['Month of year']).count()



monthly['Crash identifier'].plot.bar(figsize=(14,7), alpha=0.5, color='black', label='Total accidents')

plt.rc('xtick',labelsize=22)
plt.title('Accidents per month from 2016 to 2020')
plt.xticks(range(13))
plt.xlim(-1,12)

plt.xlabel('Month')
plt.ylabel('Accidents')
plt.legend()
```

```python

monthly = data[['Day of the week', 'Crash identifier']].groupby(['Day of the week']).count()



monthly['Crash identifier'].plot.bar(figsize=(14,7), alpha=0.5, color='b', label='Total accidents')


plt.title('Accidents per month from 2016 to 2020')

plt.rcParams.update({'font.size':22})
plt.xticks(range(7))
plt.xlim(-1,7)

plt.xlabel('Day')
plt.ylabel('Accidents')
plt.legend()
```
